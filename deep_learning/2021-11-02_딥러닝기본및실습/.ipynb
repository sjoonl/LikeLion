```python
kor = [50,70,80,90,100]
eng = [70,88,99,100,70]
math = [170,220,250,280,270]

dat = pd.DataFrame({'kor':kor, 'eng':eng, 'math':math})
dat
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = dat[['kor', 'eng']]
y = dat['math']

model = LinearRegression()
model.fit(X, y)
print("가중치. y절편 : ", model.coef_, model.intercept_)
```
