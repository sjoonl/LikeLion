```python
## train 데이터 셋 , test 데이터 셋
## train 은 학습을 위한 입력 데이터 셋
## test 은 예측을 위한 새로운 데이터 셋(평가)
## parse_dates : datetime 컬럼을 시간형으로 불러올 수 있음
train = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/2021-11-02_딥러닝기본및실습/bike_mod_tr.csv", parse_dates=['datetime'])
test = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/2021-11-02_딥러닝기본및실습/bike_mod_test.csv", parse_dates=['datetime'])
sub = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/2021-11-02_딥러닝기본및실습/sampleSubmission.csv")
```

```python
sel = ['hour', 'temp', 'workingday','windspeed', 'holiday', 'season']

X = train[sel]
y = train['count']

X_val = test[sel]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train ,y_test = train_test_split(X, y, random_state = 0)
```

#딥러닝 모델 만들기

```python
#tensorflow, pytorch, keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(128, input_dim = 6, activation = 'relu'))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(1))
```

```python
model.summary()
```

```python
model.compile(loss = "mean_squared_error", optimizer = 'rmsprop')
model.fit(X_train, y_train, epochs = 500, batch_size = 10)
```

```python
model.evaluate(X_test, y_test)
```

```python
pred = model.predict(X_val)
```

```python
sub['count'] = pred
sub.loc[sub['count']<0, 'count'] = 0
```
